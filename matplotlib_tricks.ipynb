## How to Change Plot Size?

* Use the figure() class of matplotlib for creating a new figure. 
* Pass the width and height of the desired plot in inches as the figsize argument

```python
import numpy as np
import matplotlib
import random

fig = plt.figure(figsize=(10,8))

X = list(range(10))
y = [x+(x*random.random()) for x in X]

plt.plot(X,y)

plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

## How to generate subplots?

There are three ways of generating subplots in matplotlib:

* Using subplot()
* Using subplots()
* Using subplot2grid()

```python
# subplot() – the number of rows, number of columns, and index

fig = plt.figure(figsize=(16,6))

X = list(range(10))

ax1 = plt.subplot(1,2,1)
ax1.plot(X, [x+(x*random.random()) for x in X])
ax1.set_title('Plot 1')

ax2 = plt.subplot(1,2,2)
ax2.plot(X, [x-(x*random.random()) for x in X])
ax2.set_title('Plot 2')
```

```python
fig, ax = plt.subplots(1,2, figsize=(8,6))

ax[0].plot(X, [x+(x*random.random()) for x in X])
ax[1].plot(X, [x-(x*random.random()) for x in X])

plt.show()
```

```python
fig, ax = plt.subplots(2,2, figsize=(8,6))

ax[0][0].plot(X, [x+(x*random.random()) for x in X])
ax[0][0].set_title('Plot 1')

ax[0][1].plot(X, [x-(x*random.random()) for x in X])
ax[0][1].set_title('Plot 2')

ax[1][0].plot(X, [x+(1+x*random.random()) for x in X])
ax[1][0].set_title('Plot 3')

ax[1][1].plot(X, [x*(x*random.random()) for x in X])
ax[1][1].set_title('Plot 4')


plt.show()
```

All these plots are of the same size, what if I have to modify them? 

```python
fig = plt.figure(figsize=(8,8))

ax1 = plt.subplot2grid((4,4), (0,0), colspan=3)
ax1.plot(X, [x+(x*random.random()) for x in X])
ax1.set_title('Plot 1: (0,0)')

ax2 = plt.subplot2grid((4,4),(0,3))
ax2.plot(X, [x-(x*random.random()) for x in X])
ax2.set_title('Plot 2: (0,3)')

ax3 = plt.subplot2grid((4,4),(1,0),rowspan=3,colspan=3)
ax3.plot(X, [x-(x*random.random()) for x in X])
ax3.set_title('Plot 3: (1,0)')

ax4 = plt.subplot2grid((4,4),(1,3),rowspan=3,colspan=1)
ax4.plot(X, [x+(x*random.random()) for x in X])
ax4.set_title('Plot 4: (1,3)')

fig.tight_layout()

plt.show()
```

## matplotlib Using the text() function

* Using the text() function
* Using the annotate() function of pyplot

```python
# text() takes x and y position as arguments and lets you write text there

fig = plt.figure(figsize=(8,8))

temp=[ random.uniform(20,40) for i in range(5)]
city=['City A','City B','City C','City D','City E']
y_pos=list(range(1,6))

graph = plt.bar(x=y_pos, height=temp, color='violet')

plt.xticks(y_pos, city)
plt.title('City Temperature')
plt.xlabel('Cities')
plt.ylabel('Temp')

for bar,t in zip(graph, temp):
    plt.text(bar.get_x() + bar.get_width()/2.0,bar.get_height(),'%.2f $^\circ$C'%t,ha='center')
    
plt.show()
```

```python
# To annotate a point on the plot with a marker and also write text use annotate()

fig = plt.figure(figsize=(8,8))

plt.plot(X, np.exp(X))
plt.title('Annotating Exponential Plot using plt.annotate()')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.annotate('Point 1', xy=(6,400), arrowprops=dict(arrowstyle='->'), xytext=(4,600))

plt.annotate('Point 2', xy=(7,1150), arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=-.2'), xytext=(4.5,2000))

plt.annotate('Point 3', xy=(8,3000), arrowprops=dict(arrowstyle='-|>',connectionstyle='angle,angleA=90,angleB=0'), xytext=(8.5, 2200))


plt.show()
```

## How to modify axes?

xlim() and ylim() are the two functions used for changing the limit of axes of a plot:

```python
fig = plt.figure(figsize=(8,8))

plt.plot(X, np.exp(X))
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.xlim(0,9)
plt.ylim(0,8000)


plt.annotate('Point 1', xy=(6,400), arrowprops=dict(arrowstyle='->'), xytext=(4,600))

plt.annotate('Point 2', xy=(7,1150), arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=-.2'), xytext=(4.5,2000))

plt.annotate('Point 3', xy=(8,3000), arrowprops=dict(arrowstyle='-|>',connectionstyle='angle,angleA=90,angleB=0'), xytext=(8.5, 2200))


plt.show()
```

```python
# removing axes of figureby setting spines=False
fig = plt.figure(figsize=(8,8))

plt.plot(X, np.exp(X))
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.xlim(0,9)
plt.ylim(0,8000)

#removing axes
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)

plt.annotate('Point 1', xy=(6,400), arrowprops=dict(arrowstyle='->'), xytext=(4,600))
plt.annotate('Point 2', xy=(7,1150), arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=-.2'), xytext=(4.5,2000))
plt.annotate('Point 3', xy=(8,3000), arrowprops=dict(arrowstyle='-|>',connectionstyle='angle,angleA=90,angleB=0'), xytext=(8.5, 2200))

plt.show()
```

* plt.gca() is used for accessing the current axes object.
* We can try to make it more appealing by changing the color of the axes.

```python
fig=plt.figure(figsize=(8,6))

plt.plot(X,np.exp(X))

plt.title('Annotating Exponential Plot using plt.annotate()')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

#changing axes limits
plt.ylim(1,8000)
plt.xlim(0,9)

# removing axes from the figure
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)

#changing color of the axes
plt.gca().spines['left'].set_color('red')
plt.gca().spines['bottom'].set_color('green')

plt.annotate('Point 1',xy=(6,400),arrowprops=dict(arrowstyle='->'),xytext=(4,600))

plt.annotate('Point 2',xy=(7,1150),arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=-.2'),xytext=(4.5,2000))

plt.annotate('Point 3',xy=(8,3000),arrowprops=dict(arrowstyle='-|>',connectionstyle='angle,angleA=90,angleB=0'),xytext=(8.3,2200))

plt.show()
```

## How to group bar plots?

* Bar plots are great for visualizing and comparing data among categories. Grouped bar plots allow you to do the same with different groups.
* The things you need to pay attention to here are the width and position of bars on the vertical or horizontal axis (depending on the type of bar plot, i.e., vertical or horizontal).

```python
temp_summer = [random.uniform(20,40) for i in range(5)]
temp_winter = [random.uniform(0,10) for i in range(5)]
city=['City A','City B','City C','City D','City E']

fig=plt.figure(figsize=(10,6))

x_pos_summer = list(range(1,6))
x_pos_winter = [i+0.4 for i in x_pos_summer]

graph_summer = plt.bar(x_pos_summer, height=temp_summer, label='summer', width=0.4)
graph_winter = plt.bar(x_pos_winter, height=temp_winter, label='winter', width=0.4)

plt.xticks([i+0.4 for i in x_pos_summer],city)
plt.title('City Temperature')
plt.xlabel('Cities')
plt.ylabel('Temperature ($^\circ$C)')

#Annotating Graphs
for summer_bar, winter_bar, ts, tw in zip(graph_summer, graph_winter, temp_summer, temp_winter):
    plt.text(summer_bar.get_x() + summer_bar.get_width()/2.0, summer_bar.get_height(),'%.2f $^\circ$C'%t,ha='center',va='bottom')
    plt.text(winter_bar.get_x() + winter_bar.get_width()/2.0, winter_bar.get_height(),'%.2f $^\circ$C'%t,ha='center',va='bottom')

plt.legend()
plt.show()
```

## How to modify ticks?

* Modifying ticks: xticks() and yticks()
* These functions take the position and labels as arguments and create ticks for a plot. You can use the same functions for changing the font, size, rotating ticks, and much more:These functions take the position and labels as arguments and create ticks for a plot. You can use the same functions for changing the font, size, rotating ticks.

```python
fig=plt.figure(figsize=(10,8))

temp_summer = [random.uniform(20,40) for i in range(5)]
temp_winter = [random.uniform(0,10) for i in range(5)]
city=['City A','City B','City C','City D','City E']

x_pos_summer = list(range(1,6))
x_pos_winter = [i+0.4 for i in x_pos_summer]

graph_summer = plt.bar(x_pos_summer, height=temp_summer, label='summer', width=0.4, color='tomato')
graph_winter = plt.bar(x_pos_winter, height=temp_winter, label='winter', width=0.4, color='dodgerblue')

plt.title('City Temperature')
plt.xlabel('Cities')
plt.ylabel('Temperature ($^\circ$C)')

# removing axes from figure
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)


# modifying ticks
plt.xticks([i+0.4 for i in x_pos_summer], city, fontname='Chilanka', rotation=45, fontsize=14)


#Annotating Graphs
for summer_bar, winter_bar, ts, tw in zip(graph_summer, graph_winter, temp_summer, temp_winter):
    plt.text(summer_bar.get_x() + summer_bar.get_width()/2.0, summer_bar.get_height(),'%.2f$^\circ$C'%ts,ha='center',va='bottom')
    plt.text(winter_bar.get_x() + winter_bar.get_width()/2.0, winter_bar.get_height(),'%.2f$^\circ$C'%tw,ha='center',va='bottom')

plt.legend()
plt.show()
```

## How to modify legends?

* Using the same legend() function used to place it on the axes.
* It takes various arguments loc, frameon, and ncol.
* The loc argument is used for defining the location of the legend, the frameon takes a boolean value and is used for setting the visibility of the frame around a legend and ncol is used for setting the number of columns in a legend:

```python
fig=plt.figure(figsize=(10,8))

temp_summer = [random.uniform(20,30) for i in range(5)]
temp_winter = [random.uniform(10,20) for i in range(5)]
city=['City A','City B','City C','City D','City E']

x_pos_summer = list(range(1,6))
x_pos_winter = [i+0.4 for i in x_pos_summer]

graph_summer = plt.bar(x_pos_summer, height=temp_summer, label='summer', width=0.4, color='tomato')
graph_winter = plt.bar(x_pos_winter, height=temp_winter, label='winter', width=0.4, color='dodgerblue')

plt.title('City Temperature')
plt.ylim(0,45)
plt.xlabel('Cities')
plt.ylabel('Temperature ($^\circ$C)')

# removing axes from figure
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)


# modifying ticks
plt.xticks([i+0.4 for i in x_pos_summer], city, fontname='Chilanka', rotation=45, fontsize=14)


#Annotating Graphs
for summer_bar, winter_bar, ts, tw in zip(graph_summer, graph_winter, temp_summer, temp_winter):
    plt.text(summer_bar.get_x() + summer_bar.get_width()/2.0, summer_bar.get_height(),'%.2f$^\circ$C'%ts,ha='center',va='bottom')
    plt.text(winter_bar.get_x() + winter_bar.get_width()/2.0, winter_bar.get_height(),'%.2f$^\circ$C'%tw,ha='center',va='bottom')

plt.legend(loc='upper center',ncol=2,frameon=False)
plt.show()
```
